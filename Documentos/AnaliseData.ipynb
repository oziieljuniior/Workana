### Tabela: `projetos`

Essa tabela armazena todos os projetos coletados do Workana.

| Coluna                         | Tipo          | Descrição                                           |
| ------------------------------ | ------------- | --------------------------------------------------- |
| `id`                           | INTEGER       | Identificador único do projeto no banco             |
| `hash`                         | TEXT (UNIQUE) | Hash único do projeto para evitar duplicação        |
| `titulo`                       | TEXT          | Título do projeto no Workana                        |
| `link`                         | TEXT (UNIQUE) | Link do projeto no Workana                          |
| `descricao`                    | TEXT          | Descrição completa do projeto                       |
| `cliente_verificado`           | BOOLEAN       | Indica se o cliente é verificado                    |
| `cliente_pagamento_verificado` | BOOLEAN       | Indica se o cliente tem pagamento verificado        |
| `cliente_historico`            | TEXT          | Histórico do cliente (ex: "10 projetos publicados") |
| `enviar_proposta`              | BOOLEAN       | Indica se esse projeto deve receber proposta        |
| `enviado_telegram`             | BOOLEAN       | Indica se o projeto já foi enviado ao Telegram      |
| `data_coleta`                  | TIMESTAMP     | Data e hora que o projeto foi coletado              |

```python
# Criar o filtro: 
# 1. cliente_pagamento_verificado diferente de 0 (ou seja, é 1) [cite: 133, 149]
# 2. OU cliente_historico não é uma string vazia [cite: 134, 149]
filtro_valido = (df['cliente_pagamento_verificado'] != 0) | (df['cliente_historico'] != '')

# Aplicar o filtro ao DataFrame original para manter todas as colunas [cite: 109, 110, 111]
df_final = df[filtro_valido].reset_index(drop=True)
```

## Analise da última linha

```python
def limpar_descricao(texto):
    # O conteúdo real geralmente começa após "Sobre este projeto" 
    # e termina antes de "Categoria"
    try:
        inicio = texto.split("Sobre este projeto")[1]
        fim = inicio.split("Freelancers interessados")[0]
        return fim.strip()
    except:
        return texto # Caso não encontre o padrão, mantém o original

dfLinhaTeste = limpar_descricao(df_final.iloc[-1]['descricao'])
print(dfLinhaTeste)
```
